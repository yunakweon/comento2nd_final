```python
from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt

# 모델 로드 (본인이 학습한 best.pt 경로로)
model = YOLO("/content/runs/detect/train2/weights/best.pt")

# 업로드한 이미지 경로 지정
image_path = "/content/drive/MyDrive/A22_jpg.rf.f02ad8558ce1c88213b4f83c0bc66bc8.jpg"

# 이미지 읽기 및 RGB로 변환
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#  예측
results = model(image_path)

# 탐지된 객체 그리기
for result in results:
    for box in result.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        label = result.names[int(box.cls[0])]
        confidence = float(box.conf[0])
        cv2.rectangle(image_rgb, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(image_rgb, f"{label} ({confidence:.2f})", (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# 결과 시각화
plt.figure(figsize=(10, 10))
plt.imshow(image_rgb)
plt.axis("off")
plt.title("YOLOv8 Detection Result")
plt.show()
```

```python
# 테스트 대상 함수
def translate_to_korean(word):
    return "개" if word == "DOG" else "[번역 오류]"

def class_id_to_char(cls_id):
    if cls_id < 0 or cls_id > 25:
        return "?"
    return chr(ord('A') + cls_id)

# 테스트 함수 1: 번역 테스트
def test_translate_to_korean():
    print("[translate_to_korean 테스트]")
    cases = ["DOG", "CAT", "", "HELLO"]
    for word in cases:
        result = translate_to_korean(word)
        print(f"  입력: {word:>5} → 결과: {result}")

# 테스트 함수 2: 클래스 ID 변환 테스트
def test_class_id_to_char():
    print("[class_id_to_char 테스트]")
    cases = [0, 25, -1, 26, 10]
    for cls_id in cases:
        result = class_id_to_char(cls_id)
        print(f"  입력: {cls_id:>2} → 결과: {result}")

# 테스트 실행
test_translate_to_korean()
test_class_id_to_char()

print("\n테스트 완료")
```
