```python
def translate_to_korean(word):
    return "ê°œ" if word == "DOG" else "[ë²ˆì—­ ì˜¤ë¥˜]"

def class_id_to_char(cls_id):
    if cls_id < 0 or cls_id > 25:
        return "?"
    return chr(ord('A') + cls_id)

# í…ŒìŠ¤íŠ¸
def test_translate_to_korean():
    assert translate_to_korean("DOG") == "ê°œ"
    assert translate_to_korean("CAT") == "[ë²ˆì—­ ì˜¤ë¥˜]"
    assert translate_to_korean("") == "[ë²ˆì—­ ì˜¤ë¥˜]"

def test_class_id_to_char():
    assert class_id_to_char(0) == "A"
    assert class_id_to_char(25) == "Z"
    assert class_id_to_char(-1) == "?"
    assert class_id_to_char(26) == "?"

# ì‹¤í–‰
test_translate_to_korean()
test_class_id_to_char()

print("ëª¨ë“  í…ŒìŠ¤íŠ¸ í†µê³¼!")
```
