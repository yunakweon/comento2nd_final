```python
from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt

# ëª¨ë¸ ë¡œë“œ (ë³¸ì¸ì´ í•™ìŠµí•œ best.pt ê²½ë¡œë¡œ)
model = YOLO("/content/runs/detect/train2/weights/best.pt")

# ì—…ë¡œë“œí•œ ì´ë¯¸ì§€ ê²½ë¡œ ì§€ì •
image_path = "/content/drive/MyDrive/img-421_jpg.rf.1e0bfee37e39680ac2a0f74cbd860180.jpg"

# ì´ë¯¸ì§€ ì½ê¸° ë° RGBë¡œ ë³€í™˜
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#  ì˜ˆì¸¡
results = model(image_path)

# íƒì§€ëœ ê°ì²´ ê·¸ë¦¬ê¸°
for result in results:
    for box in result.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        label = result.names[int(box.cls[0])]
        confidence = float(box.conf[0])
        cv2.rectangle(image_rgb, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(image_rgb, f"{label} ({confidence:.2f})", (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# ê²°ê³¼ ì‹œê°í™”
plt.figure(figsize=(10, 10))
plt.imshow(image_rgb)
plt.axis("off")
plt.title("YOLOv8 Detection Result")
plt.show()
```
